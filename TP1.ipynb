```python
import io
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('conn_despesas_publicas_tp1.db')
cursor = conn.cursor()
```

```python
f = io.open('despesas_publicas_tp1.sql', 'rb')
sql = f.read()
cursor.executescript(sql.decode('iso8859_1'))
```

```python
#1
df = pd.read_sql_query("SELECT S.COD_SUBFUNCAO, S.NOME_SUBFUNCAO FROM FUNCAO F NATURAL JOIN DETALHAMENTO_FUNCAO DF NATURAL JOIN SUBFUNCAO S WHERE F.COD_FUNCAO = (SELECT COD_FUNCAO FROM FUNCAO F NATURAL JOIN DETALHAMENTO_FUNCAO DF NATURAL JOIN SUBFUNCAO S GROUP BY DF.COD_FUNCAO ORDER BY COUNT(*) DESC LIMIT 1);", conn)
print(len(df.columns))
print(len(df))
```

```python
#2
df = pd.read_sql_query("SELECT F.NOME_FUNCAO, S.NOME_SUBFUNCAO, SUM(DP.VALOR_DESPESA) VT FROM DESPESA_PUBLICA DP NATURAL JOIN FUNCAO F NATURAL JOIN DETALHAMENTO_FUNCAO DF NATURAL JOIN SUBFUNCAO S GROUP BY F.COD_FUNCAO, S.COD_SUBFUNCAO HAVING VT >= 200000;",conn)
print(len(df.columns))
print(len(df))
```

```python
#3
df = pd.read_sql_query("SELECT O.NOME_ORGAO, SUM(DP.VALOR_DESPESA) VT FROM DESPESA_PUBLICA DP NATURAL JOIN ORGAO O WHERE DP.ANO_DESPESA = 2018 GROUP BY DP.COD_ORGAO ORDER BY VT DESC;",conn)
print(len(df.columns))
print(len(df))
```

```python
#6
df = pd.read_sql_query("SELECT F.NOME_FUNCAO, S.NOME_SUBFUNCAO, SUM(DP.VALOR_DESPESA) VT FROM DESPESA_PUBLICA DP NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN FUNCAO F NATURAL JOIN SUBFUNCAO S WHERE DP.COD_PROGRAMA_GOVERNO = (SELECT COD_PROGRAMA_GOVERNO FROM PROGRAMA_GOVERNO WHERE NOME_PROGRAMA_GOVERNO = \"MAIS MEDICOS\") AND DP.MES_DESPESA >= 7 AND DP.MES_DESPESA <= 9 AND DP.ANO_DESPESA = 2018;",conn)
print(len(df.columns))
print(len(df))
df
```

```python
#7
df = pd.read_sql_query("SELECT O.NOME_ORGAO, COUNT(DISTINCT PG.NOME_PROGRAMA_GOVERNO) NPROGS FROM ORGAO O NATURAL JOIN DESPESA_PUBLICA DP NATURAL JOIN PROGRAMA_GOVERNO PG WHERE DP.VALOR_DESPESA > 0 GROUP BY O.COD_ORGAO;", conn)
print(len(df.columns))
print(len(df))
df
```

```python
#10
df = pd.read_sql_query("SELECT DP.COD_DESPESA, DP.MES_DESPESA, DP.ANO_DESPESA, DP.MODALIDADE_DESPESA, DP.VALOR_DESPESA FROM DESPESA_PUBLICA DP WHERE DP.COD_ORGAO = (SELECT O.COD_ORGAO FROM DESPESA_PUBLICA DP NATURAL JOIN ORGAO O WHERE DP.ANO_DESPESA = 2018 GROUP BY DP.COD_ORGAO ORDER BY SUM(DP.VALOR_DESPESA) DESC LIMIT 1);",conn)
print(len(df.columns))
print(len(df))
df
```

```python
#12
df = pd.read_sql_query("SELECT FUNCAO.NOME_FUNCAO, SUBFUNCAO.NOME_SUBFUNCAO FROM FUNCAO NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN SUBFUNCAO;",conn)
print(len(df.columns))
print(len(df))
df.head()
```

```python
#13
df = pd.read_sql_query("SELECT DP.MODALIDADE_DESPESA, DP.VALOR_DESPESA, PG.NOME_PROGRAMA_GOVERNO FROM PROGRAMA_GOVERNO PG NATURAL JOIN DESPESA_PUBLICA DP WHERE DP.MODALIDADE_DESPESA = 'Reserva de Contingência'",conn)
print(len(df.columns))
print(len(df))
df
```

```python
#14
df = pd.read_sql_query("SELECT DP.VALOR_DESPESA, O.NOME_ORGAO FROM DESPESA_PUBLICA DP NATURAL JOIN ORGAO O WHERE O.NOME_ORGAO = 'Ministério da Fazenda';",conn)
print(len(df.columns))
print(len(df))
df
```
